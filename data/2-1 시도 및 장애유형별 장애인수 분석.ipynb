```python
# 라이브러리 호출
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
# 파일 불러오기
file_path = './정제데이터/시도 및 장애유형별 장애인수.csv'
df = pd.read_csv(file_path, index_col=0)
df
```

```python
# 시도 합계 컬럼 추가
df_sido = df.copy()
df_sido['합계'] = df_sido.sum(axis=1)
df_sido
```

```python
# 시도별 장애인수
grp_ = plt.figure(figsize=(16,10))
sns.barplot(x=df_sido.index, y='합계', data=df_sido)
plt.title('시도별 장애인수')
plt.xlabel('시도별')
plt.ylabel('장애인수')
grp_.tight_layout()
plt.show()
```

```python
'''
시도별 장애인수는 인구가 많은 경기도권과 서울특별시가 가장 많다.
'''
```

```python
# 장애유형 합계 행추가
df_type = df.copy()
df_temp = pd.DataFrame(df_type.sum()).T
df_type = pd.concat([df_type, df_temp])
df_type
```

```python
# 장애유형별 장애인수
grp_ = plt.figure(figsize=(16,10))
sns.barplot(x=df_type.columns, y=df_type.loc[0])
plt.title('장애유형별 장애인수')
plt.xlabel('장애유형별')
plt.ylabel('장애인수')
grp_.tight_layout()
plt.show()
```

```python
## 분석
# 1. 장애유형 중 지체 장애가 압도적인 비율을 차지하며, 이어 청각장애, 뇌병변 순.
# 2. 전동 보장구를 필요로 하는 장애 유형은 주로 지체장애와 뇌병변 장애
```

```python
'''
장애유형별 장애인수는 지체 장애인이 압도적으로 많다.
전동 보장구의 도움이 필요한 장애유형은 뇌병변과 지체이고
2017년 자료 기준으로 그 중에서도 실제로 몸이 불편하여 보장구가 필요한 비율은 14.7%
보유하고 있는 비율은 6.7%이다.
이 것을 인원수로 환산했을 때 보장구가 필요한 사람은 443,042명이고
전동 보장구를 소지하고 있는 사람은 201,931명이다. 
'''
```

```python
# 히트맵 작성
plt.figure(figsize=(16,10))
sns.heatmap(df, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('시도별 및 장애유형별 데이터 분석')
plt.show()
```

```python
'''
인구가 많은 서울특별시와 경기도권에서 지체 장애인의 숫자가 가장 많은 것으로 나타난다
'''
```

```python
df_type = df_type.iloc[:-1]
df_type
```

```python
df_area = list(df_type.index)
len(df_area)
```

```python
for i in df_area:
    df_temp = df_type[df_type.index == i]
    df_temp = df_temp.T.reset_index()
    
    plt.figure(figsize=(16,10))
    
    x = df_temp['index']
    y = df_temp[i]
    
    sns.barplot(x, y)
    
    plt.title('{} 장애유형별 장애인수'.format(i), fontsize=20)
    
    plt.xlabel('장애유형', fontsize = 15)
    plt.ylabel('장애인수', fontsize = 15)

    plt.grid(True)
    
    plt.savefig('./그래프/{} 장애유형별 장애인수'.format(i))
    plt.tight_layout()
    
    plt.show()
```
