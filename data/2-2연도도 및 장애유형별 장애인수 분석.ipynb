```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic', size=10)
```

```python
file_path = './원본데이터/전국_장애유형별_성별_등록장애인수_20220624103406.csv'
df = pd.read_csv(file_path, encoding='euc-kr', header=1)
df
```

```python
# 합계 삭제
df = df.iloc[1:,:]
df
```

```python
# 장애유형별 컬럼 정리
df['장애유형별(1)'][3] = '청각'
df['장애유형별(1)'][4] = '언어'
df
```

```python
# 필요없는 컬럼 삭제
df = df[['장애유형별(1)','계','계.1','계.2','계.3']]
df
```

```python
# 컬럼명 정리
df.columns = ['장애유형', '2018', '2019','2020','2021']
df
```

```python
# 그래프 그리기 위에서 데이터프레임 형태변경
a = df[['장애유형', '2018']]
b = df[['장애유형', '2019']]
c = df[['장애유형', '2020']]
d = df[['장애유형', '2021']]
```

```python
a.columns = ['장애유형', '합계']
b.columns = ['장애유형', '합계']
c.columns = ['장애유형', '합계']
d.columns = ['장애유형', '합계']
```

```python
a['년도'] = 2018
b['년도'] = 2019
c['년도'] = 2020
d['년도'] = 2021
```

```python
df_total = pd.concat([a,b,c,d])
df_total
```

```python
# 그래프 그리기
plt.figure(figsize=(15,10))
sns.barplot(x='년도', y='합계', data=df_total, hue='장애유형')
```

### 분석
### 1. 지체장애의 경우 그 빈도가 소폭 감소하는 양상을 보이나 여전히 가장 많은 비율을 차지
### 2. 지체장애와 뇌병변장애의 경우 타 장애와 다르게 이동 자체에 제한이 있어 이동권이 확보되지 않으면 기본적인 경제생활을 영위하기 힘듦 (전체 재가장애인의 60% 이상이 지체,뇌병변장애)
### * 재가장애인 : 일상생활을 하기 어려워 집에 머물며 살아가는 장애인.
### 3. 따라서 주 분석 대상을 지체,뇌병변장애로 삼았다
