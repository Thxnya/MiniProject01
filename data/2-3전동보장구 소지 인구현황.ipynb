```python
# 라이브러리 불러오기
import folium
import pandas as pd
import json
```

```python
# 파일 불러오기
sigoo = pd.read_csv('./정제데이터/시군구별 전동휠체어 필요,소지.csv')
sigoo
```

```python
# 기본지도 그리기
m=folium.Map(location=[35.256989, 127.810214],tiles='CartoDB dark_matter', zoom_start=6)

# json 파일 불러오기
gfile=open('./원본데이터/sigun.json','r',encoding='utf-8').read()
jsondata=json.loads(gfile)

# 시군별로 경계그리기
folium.GeoJson(jsondata, name='municipalities').add_to(m)

# 시군별 
folium.Choropleth(
    geo_data = jsondata,
    data = sigoo,
    columns=["시군구별", "소지"],
    fill_color = "YlGn",
    fill_opacity = 0.7,
    line_opacity = 0.7,
    key_on = "properties.SIG_KOR_NM"
).add_to(m)

m
```

```python
# 전동보장구 소지 인원이 가장 많은 10개 시
sigoo.sort_values(by='소지',ascending=False).head(10)
```
