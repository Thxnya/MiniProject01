```python
# 라이브러리 불러오기
import folium
import pandas as pd
import json
```

```python
# 파일 불러오기
sigoo = pd.read_csv('./정제데이터/시군구별 전동휠체어 필요,소지.csv')
location = pd.read_csv('./원본데이터/추가_전국전동휠체어급속충전기표준데이터.csv', encoding='euc-kr')
```

```python
# 결측치 제거
location = location.dropna()
```

```python
# 위도, 경도 컬럼만 조회해서 b에 담기
b = location[['위도','경도']]
```

### 전국 급속충전기 위치 지도

```python
# 기본지도 그리기
m=folium.Map(location=[35.256989, 127.810214],tiles='openstreetmap', zoom_start=6)

# json 파일 불러오기
gfile=open('./원본데이터/sigun.json','r',encoding='utf-8').read()
jsondata=json.loads(gfile)

# 시군별로 경계그리기
folium.GeoJson(jsondata, name='municipalities').add_to(m)

# 전동휠체어 충전소 마커찍기
for i in range(len(b)):     #index로 범위
    x=b.iloc[i,0]
    y=b.iloc[i,1]

    folium.CircleMarker(location=[x,y],color='orange',fill_color='yellow',
                    fill_opacity=0.7,weight=1.5,radius=3).add_to(m)


m
```
