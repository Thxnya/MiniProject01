```python
# 라이브러리 호출
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
file_path = './원본데이터/주된_장애유형별·성별_재가장애인_출현율_20220627140424.csv'
df = pd.read_csv(file_path, 
                 encoding='euc-kr',
                 header=1)
df = df.iloc[:,[0,2,3,4,5,6]]
df
```

```python
# 컬럼명 정리
df.columns = ['장애유형', '2005', '2008', '2011','2014','2017']
df
```

```python
# 그래프 그리기 위에서 데이터프레임 형태변경
a = df[['장애유형', '2005']]
b = df[['장애유형', '2008']]
c = df[['장애유형', '2011']]
d = df[['장애유형', '2014']]
e = df[['장애유형', '2017']]
```

```python
a.columns = ['장애유형', '합계']
b.columns = ['장애유형', '합계']
c.columns = ['장애유형', '합계']
d.columns = ['장애유형', '합계']
e.columns = ['장애유형', '합계']
```

```python
a['년도'] = 2005
b['년도'] = 2008
c['년도'] = 2011
d['년도'] = 2014
e['년도'] = 2017
```

```python
a
```

```python
a = a.sort_values(by='합계', ascending=False).head()
b = b.sort_values(by='합계', ascending=False).head()
c = c.sort_values(by='합계', ascending=False).head()
d = d.sort_values(by='합계', ascending=False).head()
e = e.sort_values(by='합계', ascending=False).head()
```

```python
df_total = pd.concat([a,b,c,d,e])
df_total
```

```python
# 연도 및 장애유형별 장애인수
grp_ = plt.figure(figsize=(16,10))
sns.barplot(x='년도', y='합계', data=df_total, hue='장애유형')
plt.title('연도 및 장애유형별 재가장애인 비율')
plt.xlabel('연도')
plt.ylabel('비율')
grp_.tight_layout()
plt.show()
```
