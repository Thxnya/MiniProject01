```python
# 라이브러리 호출
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
file_path = './정제데이터/연령 및 장애유형별 장애인수.csv'
df = pd.read_csv(file_path, index_col=0)
df
```

```python
# 나이대 합계 컬럼 추가
df_age = df.copy()
df_age['합계'] = df_age.sum(axis=1)
df_age
```

```python
# 시도별 장애인수
grp_ = plt.figure(figsize=(16,10))
sns.barplot(x=df_age.index, y='합계', data=df_age)
plt.title('나이대별 장애인수')
plt.xlabel('나이대')
plt.ylabel('장애인수')
grp_.tight_layout()
plt.show()
```

```python
# 히트맵 작성
plt.figure(figsize=(16,10))
sns.heatmap(df, annot=True, fmt='.0f', cmap='rocket_r')
plt.title('나이대별 및 장애유형별 데이터 분석')
plt.show()
```

```python
df_age = list(df.index)
len(df_age)
```

```python
for i in df_age:
    df_temp = df[df.index == i]
    df_temp = df_temp.T.reset_index()
    
    plt.figure(figsize=(16,10))
    
    x = df_temp['index']
    y = df_temp[i]
    
    sns.barplot(x, y)
    
    plt.title('{} 장애유형별 장애인수'.format(i), fontsize=20)
    
    plt.xlabel('장애유형', fontsize = 15)
    plt.ylabel('장애인수', fontsize = 15)

    plt.grid(True)
    
    plt.savefig('./그래프/{} 장애유형별 장애인수'.format(i))
    plt.tight_layout()
    
    plt.show()
```
