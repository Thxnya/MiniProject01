```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
file_path = './원본데이터/주요_장애인보조기구_필요_및_소지_현황_20220629152532.csv'
df = pd.read_csv(file_path, encoding='euc-kr', header=1)
df
```

```python
df = df.drop(columns='보조기구종류별(1)')
df.columns = [['필요 및 소지별', '2005', '2007','2011','2014','2017']]
df
```

```python
df1 = pd.DataFrame(df.iloc[0,1:] + df.iloc[1,1:])
df1['필요 및 소지별'] = '필요'
df1
```

```python
df2 = pd.DataFrame(df.iloc[2,1:] + df.iloc[3,1:])
df2['필요 및 소지별'] = '소지'
df2
```

```python
df_total = pd.concat([df1, df2])
df_total = df_total.reset_index()
df_total
```

```python
df_total.columns = ['연도', '장애인수', '필요 및 소지별']
df_total
```

```python
# 연도별 전동휠체어 필요, 소지 장애인수
grp_ = plt.figure(figsize=(16,10))
sns.barplot(x='연도', y='장애인수', data=df_total, hue='필요 및 소지별')
plt.title('연도별 전동휠체어 필요, 소지 장애인수', fontsize=20)
plt.xlabel('연도', fontsize=15)
plt.ylabel('장애인수', fontsize=15)
grp_.tight_layout()
plt.show()
```
