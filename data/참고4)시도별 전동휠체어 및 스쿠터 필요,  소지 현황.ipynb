```python
# 라이브러리 호출
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
# 파일 불러오기 및 필요한 부분만 조회
file_path = './정제데이터/시도별 전동휠체어 필요,소지.csv'
df = pd.read_csv(file_path)
df = df.set_index(df['시도별'])
df = df.iloc[:,2:]
df
```

```python
# 시도별 전동 보장구 필요 그래프
grp_ = plt.figure(figsize=(16,10))
sns.barplot(x=df.index, y='필요(명)', data=df)
plt.title('시도별 전동 보장구 필요 장애인수')
plt.xlabel('시도별')
plt.ylabel('장애인수')
grp_.tight_layout()
plt.show()
```

```python
# 시도별 전동 보장구 보유 그래프
grp_ = plt.figure(figsize=(16,10))
sns.barplot(x=df.index, y='소지(명)', data=df)
plt.title('시도별 전동 보장구 보유 장애인수')
plt.xlabel('시도별')
plt.ylabel('장애인수')
grp_.tight_layout()
plt.show()
```
